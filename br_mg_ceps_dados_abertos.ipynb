## Coleta e Tratamento de Dados
O shapefile foi gerado a partir do agrupamento de unidades, residenciais e comerciais, em função do cep e da distância dos pontos de cada *cluster* para o seu centróide, e da junção com os CEPs cuja a área corresponde ao território do municpios. Os pontos georeferenciados, foram obtidos em [OpenAddresses](https://batch.openaddresses.io/data) e os limites territórias coletados de [Portal de Mapas - IBGE](https://portaldemapas.ibge.gov.br/portal.php). A partir dos pontos fornecidos pela base e tratados pelo processo descrito, foi possível estimar um envoltório convexo que determina-se os limites de área dos CEPs que não eram equivalente a área do município.

```python
# project GeoPandas dataframe
mg_area_cep = gpd.read_file(path + 'shapefiles/br_mg_poligonos_ceps/br_mg_poligonos_ceps.shp')
mg_area_cep.drop(['layer','path', 'cidade', 'geocodigo'], axis=1, inplace=True)
mg_area_cep.to_crs(pyproj.CRS.from_epsg(4674), inplace=True)
mg_area_cep.head()
```

```python
lim_municipal = gpd.read_file(path + 'shapefiles/br_mg_limites_municipios')
lim_municipal.rename(columns={'NOME':'cidade'}, inplace=True)
lim_municipal.drop('ANODEREFER', axis=1, inplace=True)
lim_municipal.drop('LEICRIACAO', axis=1, inplace=True)
lim_municipal.head()
```

## Resultado Obtido
A visualização obtida pelas operações de tratamento e manipulação dos dados coletados, resultou no arquivo no formato shapefile disponível no [Google Drive](https://drive.google.com/drive/folders/1GLgAKU3Ejle7BIYV8CYyG_siGLirDMgB?usp=sharing) e que pode ser observado abaixo:

```python
fig, gax = plt.subplots(figsize = (20,15))
plt.title("Áreas de CEP do estado de Minas Gerais", fontsize=24)
lim_municipal.plot(ax=gax,
                   edgecolor='black',
                   linewidth=0.5)
mg_area_cep.plot(ax=gax,
                   edgecolor='black',
                   linewidth=1)
plt.axis('off')
plt.show()
```
